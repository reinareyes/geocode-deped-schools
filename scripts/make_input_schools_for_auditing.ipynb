```python
import numpy as np
import pandas as pd
import random
```

```python
random.seed(123)
```

```python
fn_out = "../input/deped_school_location_with_enrolment_2017_2018_consolidated_audit.csv"
```

```python
fn = "../input/deped_school_location_with_enrolment_2017_2018_consolidated.csv"
df = pd.read_csv(fn)
```

```python
df.head(2)
```

```python
# choose 5 random schools per DEPED_REGION with SOURCE_COORD = "GMAP" AND GMAP_DELTA_KM < 100 KM
regions = df["DEPED_REGION"].unique()
regions
```

```python
nsamp = 1
df_out = pd.DataFrame()
for region in regions:
    df_sub = df[(df["DEPED_REGION"]==region) & ((df["SOURCE_COORD"]=="GMAP") & (df["GMAP_DELTA_KM"]<100))]
    index_samp = random.sample(df_sub.index, nsamp)
    df_out = pd.concat([df_out, df_sub.loc[index_samp]])
```

```python
df_out.to_csv(fn_out, index=False)
```
